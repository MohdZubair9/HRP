```python
import pandas as pd 
df = pd.read_excel('hrp.xlsx')
import pandas as pd

df.columns = df.iloc[0]
df = df.drop(index=0)
# Rename Bengali columns to English
df = df.rename(columns={
    'গর্ভকাল': 'Gestational Age',
    'ওজন': 'Weight',
    'উচ্চতা': 'Height',
    'রক্ত চাপ': 'Blood Pressure',
    'রক্তস্বল্পতা': 'Anemia',
    'জন্ডিস': 'Jaundice',
    'গর্ভস্হ শিশু অবস্থান': 'Fetal Position',
    'গর্ভস্হ শিশু নাড়াচাড়া': 'Fetal Movement',
    'গর্ভস্হ শিশু হৃৎস্পন্দন': 'Fetal Heartbeat',
    'প্রসাব পরিক্ষা এলবুমিন': 'Urine Albumin',
    'প্রসাব পরিক্ষা সুগার': 'Urine Sugar',
    'ঝুকিপূর্ণ গর্ভ': 'High Risk Pregnancy',
})

# df.columns= ['pregnancy time','weight','height','Blood Pressure','Anemia','Jaundice','Position of baby','Movement of baby in womb','Baby Heartbeat','Albomine','Urine Test sugar','VDRL','HRsAG',"HRP"]
df = df.reset_index(drop=True)
df.to_excel('hrp_cleaned.xlsx', index=False)


df.head()

```

```python
# df.info()
df.isnull().sum()
```

```python
df = df.drop(columns=['Name'])
# df.head()
df['Gravida'] = df['Gravida'].str.replace('st','').str.replace('nd','').str.replace('rd','').str.replace('th','')
df.head()
```

```python
df['Gravida'] = df['Gravida'].str.replace('st','').str.replace('nd','').str.replace('rd','').str.replace('th','')
df.head(20)
```

```python

df['Weight'] = df['Weight'].str.replace(' kg', '').astype(int)

# Convert height from feet-inches string to cm
df['Height'] = df['Height'].str.replace("''", '').astype(float)
df['Height_cm'] = df['Height'] * 30.48  # approximate conversion

df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)


df['Fetal Heartbeat'] = df['Fetal Heartbeat'].str.replace('m', '').astype(int)

df['Gestational Age'] = df['Gestational Age'].str.replace(' week', '').astype(int)
df.head(20)
```

```python
df['TiTi Tika'] = df['TiTi Tika'].str.replace('st', '').str.replace('nd', '').str.replace('rd', '').astype(int)
df.head(20)
```

```python
# df.head()
# df['Anemia'].unique()
df['Anemia'].fillna(0, inplace=True)
```

```python
df['Jaundice'].fillna(0, inplace=True)
# df.head(20)
# df['Jaundice'].unique()

```

```python
for col in ['Jaundice','Anemia']:
    df[col] = df[col].replace({'Minimal':1,'Medium':2})
df.head(20)    
```

```python
# df['Fetal Movement'].unique()
# df['Fetal Position'].unique()
for col in ['Fetal Movement', 'Fetal Position']:
    df[col] = df[col].replace({'Abnormal': 1, 'Normal': 0})
df.head(20)    
```

```python
df['Fetal Movement'] = df['Fetal Movement'].astype(int)
df['Fetal position'] = df['Fetal Position'].astype(int)
```

```python
df.head(20)
```

```python
for col in ['Urine Albumin', 'Urine Sugar']:
    print(df[col].unique(),col)
```

```python
df['Urine Albumin'].fillna(0, inplace=True)
```

```python
# df['Urine Albumin'].unique()
df['Urine Albumin'] = df['Urine Albumin'].replace({'Higher':3, 'Medium':2, 'Minimal':1})
df['Urine Albumin'] = df['Urine Albumin'].astype(int)
df.head()
```

```python
df['Urine Sugar'].unique()
df['Urine Sugar']= df['Urine Sugar'].replace({"Yes":1, "No":0}, inplace=True)
df.head()
```

```python
df['Urine Sugar'].nunique()
# df.head(20)
```

```python
for col in ['VDRL', 'HRsAG']:
    print(df[col].unique()) 
    
```

```python
for col in ['VDRL', 'HRsAG']:
    df[col] = df[col].replace({'Positive': 1, 'Negative': 0})
    df[col] = df[col].astype(int)
df.head(20)    
```

```python
df['High Risk Pregnancy']= df['High Risk Pregnancy'].replace({'Yes': 1, 'No': 0})
df['High Risk Pregnancy'] = df['High Risk Pregnancy'].astype(int)
df.head()
```

```python
df.to_excel('hrp_final.xlsx', index=False)
```
